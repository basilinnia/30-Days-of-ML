```python
import tensorflow as tf
import matplotlib.pyplot as plt

#define the model
model = tf.keras.Sequential([
    # units=1: This specifies the number of neurons in the layer
    # input_shape=[1]: This specifies the expected input shape for this layer (one dimension)
    tf.keras.layers.Dense(1, input_shape=[1]),
])

# This line creates a NumPy array named x containing
# 100 equally spaced values between -1.0 and 1.0 (inclusive). 
# This represents the input data points for your model.
x = tf.linspace(-1.0, 1.0, 100)

#compile the model
y = model.predict(x)

#plot the model
plt.figure(dpi=100)
plt.plot(x, y, 'k')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("Input: x")
plt.ylabel("Target y")

w, b = model.weights # you could also use model.get_weights() here
plt.title("Weight: {:0.2f}\nBias: {:0.2f}".format(w[0][0], b[0]))
plt.show()
```

## But what is a Sequantial Model?

A sequential model is a model that has a linear stack of layers. Data flows through these layers in a single direction, from the first layer (input) to the last layer (output).

 In Keras, the alternative of the sequential model is the functional model. Functional API allows you to have more than one outputs or inputs in your model. 

So, to summarise, the Sequential model is a simple and straightforward way to define a linear stack of layers, while the Functional model is more flexible and powerful, allowing for more complex models with multiple inputs or outputs, shared layers, and non-linear flows.
